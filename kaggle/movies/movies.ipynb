```python
# Импортируем необходимые модули
import pandas as pd
```

```python
# Считаем данные и посмотрим на их вид
movies = pd.read_csv("movies.csv")
movies.head()
```

```python
# Посмотрим на типы и описательную статистику
movies.dtypes
```

```python
movies.describe()
```

```python
# Сохраним ссылку на фильм и постер в отдельную переменную, чтоб эти колонки не мешались.
# При необходимости можно будет найти нужную ссылку через индекс
urls_movies_and_posters = movies[['Movie URL','Poster']].copy()
urls_movies_and_posters.head()
```

```python
# Удалим скопированные колонки из начального датафрейма
movies.drop(['Movie URL','Poster'], axis=1, inplace=True)
movies.head()
```

```python
# Заменим символ | на запятую с пробелом. При этом, если данный символ находится в конце строки - удалим его.
movies['Directors'] = movies['Directors'].str.replace('|', ', ')
movies['Writers']   = movies['Writers'].str.replace('|', ', ')
movies['Stars']     = movies['Stars'].str.replace('|', ', ')
movies['Genres']    = movies['Genres'].str.replace('|', ', ')
```

```python
def del_last_comma(lines):
    if str(lines)[-1] == ' ':
        lines = str(lines)[:-2:]
    return lines
```

```python
movies.Directors = movies.Directors.apply(del_last_comma)
movies.Writers   = movies.Writers.apply(del_last_comma)
movies.Stars     = movies.Stars.apply(del_last_comma)
movies.Genres    = movies.Genres.apply(del_last_comma)

movies.head()
```

```python
# Избавимся от строк с None
movies.isna().sum()
```

```python
movies = movies.dropna()
movies.isna().sum()
```

```python
movies.head()
```

```python
# Переведем колонки в "Release Year", "Rating Count" и "Length in Min" в формат int
movies['Release Year']  = movies['Release Year'].apply(lambda x: int(x))
movies['Length in Min'] = movies['Length in Min'].apply(lambda x: int(x))
movies['Rating Count']  = movies['Rating Count'].apply(lambda x: int(x))

movies.head()
```

*3. Какого жанра больше всего?*

```python
# Посмотрим какие жанры у нас есть:
ganres = set()
def get_ganres(lines):
    global ganres
    gen = lines.split(', ')
    for g in gen:
        ganres.add(g)
```

```python
movies['Genres'].apply(get_ganres)
```

```python
ganres
```

```python
# Посмотрим какого жанра больше всего:
rating_ganres = {}
for ganre in ganres:
    rating_ganres[ganre] = 0

def count_rating_ganres(lines):
    global rating_ganres
    ganre_movie = lines.split(', ')

    for g in ganre_movie:
        rating_ganres[g] += 1
```

```python
movies['Genres'].apply(count_rating_ganres)
```

```python
rating_ganres
```

```python
top_ganres = pd.Series(rating_ganres)
top_ganre  = top_ganres.idxmax()
```

```python
top_ganre
```

*4. Какой актер встречается чаще всего?* <br>

```python
stars = set()
def get_stars(lines):
    global stars
    star = lines.split(', ')
    for s in star:
        stars.add(s)

movies['Stars'].apply(get_stars)

stars
```

```python
rating_stars = {}
for star in stars:
    rating_stars[star] = 0

def count_rating_stars(lines):
    global rating_stars
    star_ = lines.split(', ')

    for s in star_:
        rating_stars[s] += 1

movies['Stars'].apply(count_rating_stars)

rating_stars
```

```python
top_stars = pd.Series(rating_stars)
top_star  = top_stars.idxmax()

top_star
```
